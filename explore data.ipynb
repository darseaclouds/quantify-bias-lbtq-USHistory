```python
def euclideanDistance(v1, v2):
    n1 = np.array(v1)
    n2 = np.array(v2)
    
    return (sum(np.square(n1-n2))) ** 0.5
```

```python
import sys
MAX_INT = sys.maxsize

def cosineSimilarity(v1, v2):
    n1 = np.array(v1)
    n2 = np.array(v2)
    
    dot = sum(n1*n2)
    n1mag = sum(n1**2) ** 0.5
    n2mag = sum(n2**2) ** 0.5
    
    if n1mag*n2mag <= 0.0 or dot <=0:
        return 1-MAX_INT
    
    return 1-(dot / (n1mag*n2mag))
```

```python
def findNearestNeighbors(vec, dataframe, metric, k):
    assert metric == 'euclidean' or metric == 'cosine'
    
    if metric == 'euclidean':
        distances = [(euclideanDistance(vec, list(row)), index) for index,row in dataframe.iterrows()]
        return [v for k,v in sorted(distances)[:k]]
        
    else:
        distances = [(cosineSimilarity(vec, list(row)), index) for index,row in dataframe.iterrows()]
        #print(distances)
        return [v for k, v in sorted(distances)[(-1)*k:]]
```
